Deep Learning is a subset of machine learning, which itself is a branch of artificial intelligence (AI). It focuses on teaching computers to learn and make decisions by mimicking the way the human brain works. It uses artificial neural networks (ANNs) to process data and make predictions or classifications. These networks consist of layers of interconnected nodes (neurons), where each layer learns to extract and process different features from the input data.
Deep learning is particularly powerful for tasks where feature extraction is challenging or where data is unstructured, such as images, audio, and text. It has enabled significant advancements in fields like natural language processing (NLP), computer vision, and speech recognition.

## Differences between Machine Learning and Deep Learning

| **Aspect**               | **Machine Learning (ML)**                              | **Deep Learning (DL)**                             |
|--------------------------|-------------------------------------------------------|--------------------------------------------------|
| **Definition**           | A subset of AI focused on enabling machines to learn from data using algorithms. | A subset of ML that uses neural networks with multiple layers to model complex patterns. |
| **Data Dependency**      | Performs well with smaller datasets.                  | Requires large amounts of labeled data to perform effectively. |
| **Feature Engineering**  | Relies on manual feature extraction and selection.     | Automatically extracts features through multiple layers of the network. |
| **Model Complexity**     | Models are simpler (e.g., linear regression, SVM).     | Models are complex with deep neural network architectures. |
| **Computation Power**    | Requires less computational power.                     | Requires significant computational resources (GPUs/TPUs). |
| **Training Time**        | Faster training times for most models.                 | Training can be time-consuming due to large datasets and model complexity. |
| **Interpretability**     | Easier to interpret and understand model decisions.    | Often considered a "black box" with lower interpretability. |
| **Applications**         | Suitable for tabular data (e.g., fraud detection, customer segmentation). | Excels in unstructured data like images, text, audio (e.g., image recognition, NLP). |
| **Scalability**          | Limited scalability with increasing data complexity.   | Highly scalable for large and complex datasets. |
| **Example Algorithms**   | Linear Regression, Decision Trees, Random Forests, SVM. | Convolutional Neural Networks (CNNs), Recurrent Neural Networks (RNNs), Transformers. |

## Neural Network

### Neuron

A neuron in deep learning is the basic building block of artificial neural networks, inspired by the biological neurons in the human brain. It is a computational unit that takes inputs, processes them, and produces an output.

#### Components of a Neuron

1. Inputs ($ 𝑥_1, x_2, ... , x_n $):
    - These represent features or data points.
    - For example, in image classification, they could be pixel values.
2. Weights ($ w_1, w_2, ... , w_n $):
    - Each input has an associated weight that signifies its importance.
    - Weights are learned during training.
3. Bias ($ b $):
    - Bias helps the model shift the activation function, improving learning capabilities.
    - It acts as an offset.
4. Summation Function:  
    The neuron computes a weighted sum of inputs:

$$
 z = \sum_{i=1}^{n} w_i \cdot x_i + b
$$

5. Activation Function 
    - This introduces non-linearity, allowing the network to learn complex patterns.
    - Common activation functions include:
        - Sigmoid: Outputs values between 0 and 1.
        - ReLU (Rectified Linear Unit): Outputs max(0, z)
        - Tanh: Outputs values between -1 and 1.
    - Without activation functions, the neuron would act as a simple linear equation. This limits the network to learning only linear relationships. Activation functions introduce non-linearity, enabling the network to capture more complex patterns.
6. Output:  
   The final output is:

$$
y = f(z)
$$

#### Pre Activation Value

In a neural network, **pre-activation** is the value computed **before applying the activation function**.

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

This **(z)** is called the **pre-activation value**

**Why we need the sum**

A neuron is designed to produce one output value from multiple inputs. To do that, it must combine all input signals into a single number. The sum is the simplest and most effective way to do this.

It’s the most general and efficient way to:

- Represent linear relationships
- Scale contributions of each feature
- Learn patterns through optimization

In fact, this is the same idea used in:

- Linear regression
- Logistic regression

Neural networks are just stacked versions of this idea.

**Without weights?**

If you just added inputs:

$$
x1+x2+x3
$$

Every feature would be equally important. Model couldn’t learn which features matter more

**Decision Score**

Each input gives evidence, and the neuron:

- weighs each piece of evidence → $w_i * x_i$
- then adds them together → total evidence

$$
z = w_1*x_1+w_2*x_2+w_3*x_3+b
$$

This sum becomes a single score used for the final decision.

**Why bias is important**

1. Allows non-zero output when input is zero  
    Without bias:
    ```
    input = 0
    wx = 0
    ```
    So every neuron output becomes fixed by the activation at zero.

    With bias:
    ```
    wx + b = b
    ```
    The neuron can still learn meaningful behavior.

    Example: predicting house price:

    A house with 0 size still has a base value (land cost). So output should NOT be forced to 0 when input is 0

2. Shifts the decision boundary  
    Bias acts like the intercept in linear equations.  
    Without bias:
    ```
    y=2x
    ```
    This line must pass through the origin (0,0).  
    With bias:
    ```
    y=2x+3
    ```
    Now the line can move up/down. Neural networks need this flexibility to fit real data.

4. Helps learn complex patterns  
    * Without bias, the network becomes too constrained.
    * Many functions cannot be approximated well because every neuron is forced into limited shapes.
    * Bias gives neurons freedom to activate at different thresholds.

#### Activation Function

After computing the weighted sum (z), we pass it through an **activation function**:

$$
a = f(z)
$$

**The critical reason: Non-linearity**

If you don’t use activation functions:

* Every layer just does linear transformation
* Stacking layers = still linear

A deep network would behave like a **single-layer linear model**

If a neural network has **no activation functions**, every layer just does:

$$
z = Wx + b
$$

Stack multiple layers:

$$
z_2 = W_2(W_1x + b_1) + b_2
$$

This simplifies to:

$$
z = Wx + b
$$


> No matter how deep the network is → it cannot learn complex patterns

**With activation functions**

You introduce **non-linearity**, allowing the model to learn:

* Curves
* Complex decision boundaries
* Real-world patterns (images, text, speech)


Activation functions (like ReLU, sigmoid) introduce **non-linearity**:

$$
a = f(z)
$$

Now the model becomes:

$$
a = f(Wx + b)
$$

Stack layers:

$$
a_2 = f(W_2 \cdot f(W_1x + b_1) + b_2)
$$

This is **no longer linear**

**What can a linear model learn?**

Only straight-line relationships.

Example:

- In 2D → a straight line
- In 3D → a flat plane

It cannot learn:

- Curves
- Circles
- Complex shapes


| Input (x) | Linear Output ($z = 2x-3$) | ReLU Output ( $\max(0,z)$ ) |
| --------- | ------------------------ | ------------------------- |
| 0         | -3                       | 0                         |
| 1         | -1                       | 0                         |
| 2         | 1                        | 1                         |
| 3         | 3                        | 3                         |
| 4         | 5                        | 5                         |

 

````python
import numpy as np
import matplotlib.pyplot as plt

# Input values
x = np.linspace(-1, 5, 200)

# Linear function
z = 2 * x - 3

# ReLU activation
relu = np.maximum(0, z)

# Create plots
plt.figure(figsize=(10, 5))

# Linear plot
plt.subplot(1, 2, 1)
plt.plot(x, z, label='y = 2x - 3', color='blue')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# ReLU plot
plt.subplot(1, 2, 2)
plt.plot(x, relu, label='ReLU(2x - 3)', color='red')
plt.title('Nonlinear Function with ReLU')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
````

### Architecture of Neural Network

A neural network is a computational model designed to simulate the way the human brain processes information. It consists of layers of interconnected nodes (neurons), organized into an input layer, one or more hidden layers, and an output layer.

#### Components of a Neural Network

1. Input Layer:
    - Accepts raw data as input. Each neuron in this layer corresponds to a feature in the input data.
    - Example: If you are predicting house prices, features like "size", "location", and "number of rooms" will be inputs.
2. Hidden Layers:
    - Process data using weights, biases, and activation functions to learn intermediate representations.
    - The number of layers and neurons depends on the complexity of the task.
    - The number of node in a hidden layer can be determined by the number of feature required to extract.
3. Output Layer:
    - Provides the final result of the model.
    - Example: In classification, this layer outputs probabilities or class labels.
4. Weights and Biases:
    - Weights determine the importance of a connection between neurons.
    - Bias adjusts the weighted sum output for flexibility in learning.
5. Activation Function:
    - Introduces non-linearity, enabling the network to solve complex problems.
    - Common activation functions: Sigmoid, ReLU, Tanh, Softmax.

#### How a Neural Network Works: Step-by-Step

1. Forward Propagation:
    - Data flows through the network from the input layer to the output layer.
    - Each neuron computes the weighted sum of inputs, applies an activation function, and sends the output to the next layer.
2. Loss Calculation:
    - A loss function measures the error between the predicted and actual values.
3. Backpropagation:
    - The network adjusts weights and biases to minimize the error.
    - This involves calculating gradients using the chain rule and updating parameters with an optimization algorithm like gradient descent.
4. Training:
    - The process of forward propagation, loss calculation, and backpropagation repeats over multiple iterations (epochs) until the network learns the patterns in the data.

## Loss Function

A loss function in deep learning quantifies how well the predicted outputs of a neural network match the actual target values. It is a critical component that guides the optimization process during training by measuring the error or deviation between predictions and ground truths. The goal of training a neural network is to minimize this loss.

### Types

Loss functions can be categorized based on the type of task:

#### 1. Regression Loss Functions

**MSE**

$$
L = \frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2
$$

**MAE**

$$
L = \frac{1}{n}\sum |y_i-\hat{y}_i|
$$

#### 2. Classification Loss Functions
**Binary Cross-Entropy:**

$$
\text{BCE} = - \frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$$
   
**Categorical Cross-Entropy:**

$$
\text{CCE} = - \frac{1}{N} \sum_{i=1}^{N} \sum_{j=1}^{C} y_{ij} \log(\hat{y}_{ij})
$$
   
**Hinge Loss:**

$$
\text{Hinge Loss} = \frac{1}{N} \sum_{i=1}^{N} \max(0, 1 - y_i \cdot \hat{y}_i)
$$

### Choosing a Loss Function
**1. Regression**:
- **MSE** for smoother models that penalize large errors.
- **MAE** for robust models against outliers.

**2. Classification**:
- **Binary Cross-Entropy** for binary problems.
- **Categorical Cross-Entropy** for multi-class problems.
- **Hinge Loss** for margin-based classifiers like SVMs.

**3. Custom Tasks**:
- Combine loss functions or define your own based on specific requirements.

## Activation Function

An activation function in a neural network defines the output of a node (neuron) given an input or a set of inputs. It determines whether a neuron should be activated or not based on the input it receives. Activation functions introduce non-linearity into the model, enabling it to learn and perform complex tasks.

Without activation functions, neural networks would behave like a linear regression model, unable to capture the non-linear patterns in data.


### Types

#### 1. Linear Activation Function
- Formula: 𝑓(𝑥) = 𝑥
- Problem: Does not introduce non-linearity; all layers collapse into a single layer.
#### 2. Non-Linear Activation Functions

These are the most commonly used.

- **Sigmoid**:
    - Formula: $ \frac{1}{1 + e^{-x}} $
    - Range: 0 to 1
    - Usage: Good for binary classification; used in the output layer.
    - Drawbacks: Can cause vanishing gradients.
- **Tanh (Hyperbolic Tangent)**:
    - Formula: 𝑓(𝑥) = tanh(𝑥) = $ \frac{e^x - e^{-x}}{e^x + e^{-x}} $
    - Range: −1 to 1
    - Usage: Good for hidden layers; addresses vanishing gradients better than sigmoid.
    - Drawbacks: Still susceptible to vanishing gradients for very large or small inputs.
- **ReLU (Rectified Linear Unit)**:
    - Formula: f(x)=max(0,x)
    - Range: 0 to ∞
    - Usage: Widely used in hidden layers; fast to compute.
    - Drawbacks: Can suffer from "dead neurons" where gradients become zero for inputs less than 0.
- **Leaky ReLU**:
    - Formula: f(x)=x if x>0, otherwise f(x)=αx(α>0)
    - Usage: Solves the "dead neurons" issue by allowing a small gradient for negative inputs.
- **Softmax**:
    - Formula: $ \frac{e^{x_i}}{\sum_{j} e^{x_j}} $
    - Range: 0 to 1
    - Usage: Used in the output layer for multi-class classification; ensures probabilities sum to 1.

### Example

Suppose we have a neural network layer receiving the input vector $x = [1, -2, 3]$ and the weights $w = [0.5, 0.1, -0.4]$ with a bias $b = 0.2$.

#### Step 1: Compute the pre-activation output (z):

$$
z = w \cdot x + b
$$

$$
z = (1 \times 0.5) + (-2 \times 0.1) + (3 \times -0.4) + 0.2
$$

$$
z = 0.5 - 0.2 - 1.2 + 0.2 = -0.7
$$

#### Step 2: Apply an activation function.

- **Sigmoid**: $f(z) = \frac{1}{1 + e^{0.7}} \approx 0.332$
- **ReLU**: $f(z) = \max(0, -0.7) = 0$
- **Tanh**: $f(z) = \tanh(-0.7) \approx -0.604$

#### Explanation of Outputs:

- **Sigmoid** squashes the value to a range between 0 and 1, indicating activation strength.
- **ReLU** outputs 0, meaning the neuron is inactive for negative input.
- **Tanh** outputs a negative value, showing activation while preserving sign.

### Choosing an Activation Function

- **Hidden Layers**:
  - **ReLU**: Default choice due to simplicity and efficiency.
  - **Leaky ReLU**: Use if the dead neuron issue arises.
- **Output Layer**:
  - **Sigmoid**: For binary classification.
  - **Softmax**: For multi-class classification.
  - **Linear**: For regression problems.

## Optimizer

An optimizer updates the model’s weights so the loss decreases.

In simple terms:

- Loss tells us how wrong the model is
- Optimizer decides how to adjust weights to reduce that error

**Core Idea**

Training means finding weights $ w $ that minimize the loss:

$$
\text{Loss}(w)
$$

Optimizers usually do this using Gradient Descent.

### How Optimizers Work

#### 1. Forward Pass

- Input → model → prediction
- Compute loss

#### 2. Backward Pass (Backpropagation)

##### Output Layer Delta Term

We compute:

$$
\delta=\frac{\partial L}{\partial z}
$$

This is called the error term (delta).

It measures:

> how much changes in $ z $ affect the loss

**Dependency Chain**

$$
w,b \rightarrow z \rightarrow \hat y \rightarrow L
$$

**Apply Chain Rule**

$$
\frac{\partial L}{\partial z}
=
\frac{\partial L}{\partial \hat y}
\cdot
\frac{\partial \hat y}{\partial z}
$$

**Derivative of Loss**

Using Mean Squared Error:

$$
L=\frac12(y-\hat y)^2
$$

Then:

$$
\frac{\partial L}{\partial \hat y}
=
(\hat y-y)
$$

**Derivative of Sigmoid**

Sigmoid:

$$
\sigma(z)=\frac{1}{1+e^{-z}}
$$

Derivative:

$$
\sigma'(z)=\sigma(z)(1-\sigma(z))
$$

Since:

$$
\hat y=\sigma(z)
$$

then:

$$
\frac{\partial \hat y}{\partial z}
=
\hat y(1-\hat y)
$$

**Combine Everything**

$$
\delta
=
(\hat y-y)\hat y(1-\hat y)
$$

This is the output layer error term.

##### Hidden Layer Delta Term

For a hidden neuron $ j $, we define:

$$
\delta_j = \frac{\partial L}{\partial z_j}
$$

We want to compute how much the loss changes when $ z_j $ changes.

But there is a problem:

- $ z_j $ does NOT directly affect the loss
- it affects the next layer first

So we must go **through the next layer**

**Understand the forward flow**

For hidden neuron $ j $:

$$
z_j \rightarrow a_j \rightarrow z_k \rightarrow L
$$

So influence path is:

$$
z_j → a_j → z_k → L
$$

where:
- $ z_j $: current layer's node
- $ a_j = f(z_j) $: activation function on current layer's node
- $ z_k = \sum_j w_{jk} a_j + b_k $: next layer node
- $ L $: computed loss after predication

**Apply chain rule**

We expand:

$$
\frac{\partial L}{\partial z_j}
=
\sum_k
\frac{\partial L}{\partial z_k}
\cdot
\frac{\partial z_k}{\partial a_j}
\cdot
\frac{\partial a_j}{\partial z_j}
$$

We sum over all neurons $ k $ in the next layer because: hidden neuron $ j $ affects ALL of them

> The summation appears because: chain rule applies per path, and hidden neurons have multiple outgoing paths

**Identify each term**

First term

$$
\frac{\partial L}{\partial z_k} = \delta_k
$$

This is definition of delta at next layer.

Second term

Recall:

$$
z_k = \sum_j w_{jk} a_j + b_k
$$

So:

$$
\frac{\partial z_k}{\partial a_j} = w_{jk}
$$

Meaning:
> how much neuron $ j $ influences neuron $ k $

Third term

Since:

$$
a_j = f(z_j)
$$

Then:

$$
\frac{\partial a_j}{\partial z_j} = f'(z_j)
$$

**Substitute everything**

Now plug into chain rule:

$$
\frac{\partial L}{\partial z_j}
=
\sum_k
\delta_k \cdot w_{jk} \cdot f'(z_j)
$$

Factor out $ f'(z_j) $(constant):

$$
\delta_j =
\left(\sum_k \delta_k w_{jk}\right)\cdot f'(z_j)
$$

**Intuition**

This equation has 2 parts:

1. Future error contribution

$$
\sum_k \delta_k w_{jk}
$$

Meaning:  
> “How much did neuron $ j $ contribute to errors in the next layer?”

- if weights are large → more responsibility
- if next layer errors are large → more blame

2. Local sensitivity

$$
f'(z_j)
$$

Meaning:

> “How sensitive is neuron $ j $ to changes in its input?”

- sigmoid near extremes → gradient small
- ReLU → either 0 or 1

**Final Intuition in One Line**

Hidden layer delta =

> (error coming from next layer) × (how strongly this neuron affects them) × (how sensitive this neuron is)

**Why Output Layer is Different**

For output neuron:

- we compare directly with target $ y $

So:

$$
\delta = \frac{\partial L}{\partial z}
$$

But for hidden layers:

- no direct loss
- must propagate error backward

**Summary Flow**

```
Output layer:
δ = direct loss gradient

Hidden layer:
δ = (weighted sum of next layer δ) × local derivative
```

##### Gradient for Weights

Using chain rule:

$$
\frac{\partial L}{\partial w}
=
\frac{\partial L}{\partial \hat y}
\cdot
\frac{\partial \hat y}{\partial z}
\cdot
\frac{\partial z}{\partial w}
$$

Group first two terms:

$$
\delta
=
\frac{\partial L}{\partial z}
$$

Therefore:

$$
\frac{\partial L}{\partial w}
=
\delta
\cdot
\frac{\partial z}{\partial w}
$$

Since:

$$
z=wx+b
$$

then:

$$
\frac{\partial z}{\partial w}=x
$$

because:

- derivative of $ wx $ with respect to $ w $ is $ x $
- derivative of $ b $ is 0

Therefore:

$$
\frac{\partial L}{\partial w}
=
\delta x
$$

##### Gradient for Bias

Since:

$$
z=wx+b
$$

then:

$$
\frac{\partial z}{\partial b}=1
$$

Using chain rule:

$$
\frac{\partial L}{\partial b}
=
\frac{\partial L}{\partial z}
\cdot
\frac{\partial z}{\partial b}
$$

Therefore:

$$
\frac{\partial L}{\partial b}
=
\delta
$$

#### 3. Parameter Update

Update weights using Gradient Descent:

$$
w=w-\eta\frac{\partial L}{\partial w}
$$

Since:

$$
\frac{\partial L}{\partial w}=\delta x
$$

then:

$$
w=w-\eta(\delta x)
$$

Update bias using Gradient Descent:

$$
w=w-\eta\frac{\partial L}{\partial b}
$$

Since:

$$
\frac{\partial L}{\partial b}=\delta
$$

then:

$$
b=b-\eta\delta
$$

#### 4. Intuition

Imagine walking downhill in fog trying to reach the lowest point.

- Loss → height of the hill
- Gradient → slope direction
- Optimizer → how you choose your steps

The optimizer repeatedly:

1. computes error
2. computes gradients
3. updates parameters
4. moves toward lower loss

#### 5. Types of Optimizers 

1. **SGD (Stochastic Gradient Descent)**

    Updates weights using one (or few) samples:

    $$
    W = W - \eta \cdot \nabla L
    $$

    - Uses a fixed learning rate for updates.
    - Training may be slower and prone to oscillations.
    - simple and effective for convex problems but can struggle with complex deep learning models.

    **Pros:**
    * Simple
    * Works well for large data

    **Cons:**
    * Noisy updates
    * Slow convergence

2. **SGD with Momentum**

    Adds memory of past gradients:
    $$
    v = \gamma v + \eta \nabla L
    $$
    $$
    W = W - v
    $$
    Think: rolling ball gaining speed

    **Benefits:**
    * Faster convergence
    * Reduces oscillation

3. **Adagrad**

    Adjusts learning rate per parameter:

    $$
    W = W - \frac{\eta}{\sqrt{G + \epsilon}} \cdot \nabla L
    $$

    Frequently updated weights → smaller learning rate

    **Problem:** Learning rate keeps shrinking too much

4. **RMSprop**

    - Adapts learning rates for each parameter.
    - Handles non-stationary objectives well, leading to faster convergence.
    - adapts learning rates dynamically, making it suitable for deep neural networks.

    Fixes Adagrad by using moving average:

    $$
    E[g^2] = \beta E[g^2] + (1-\beta)g^2
    $$

    **Benefits:**

    * Stable learning
    * Works well for RNNs

5. **Adam**

    - Combines momentum and adaptive learning rates.
    - Typically achieves the best results with minimal hyperparameter tuning.
    - versatile and widely used optimizer that combines the strengths of SGD and RMSprop, making it the default choice for most tasks.

    Combines:

    * Momentum (mean of gradients)
    * RMSprop (variance of gradients)

    $$
    m = \beta_1 m + (1-\beta_1)g
    $$
    $$
    v = \beta_2 v + (1-\beta_2)g^2
    $$

    Why Adam is popular:

    * Adaptive learning rate
    * Fast
    * Works well in most cases

5. **Others (advanced)**

    * AdamW (better regularization)
    * Nadam (Adam + Nesterov momentum)
    * LAMB (large-scale models)
    * Lion (recent optimizer)

#### Key Concepts in Optimizers

**Learning Rate (η)**

* Too high → overshoot
* Too low → slow learning

Typical Learning Rate Values

| Optimizer | Common Learning Rate |
| --------- | -------------------- |
| SGD       | 0.01                 |
| Adam      | 0.001                |
| RMSProp   | 0.001                |


**Gradient**

* Direction of steepest increase
* We go **opposite** direction

**Convergence**

* When loss stops decreasing

````python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Features:
# x1 = Hours Studied
# x2 = Attendance

X = np.array([
    [2, 1],
    [3, 2],
    [7, 6],
    [8, 7]
], dtype=float)

# Target:
# 0 = Fail
# 1 = Pass

y = np.array([
    [0],
    [0],
    [1],
    [1]
], dtype=float)

# Normalize Data: Scaling helps neural networks train better

X = X / np.max(X, axis=0)

model = Sequential([
    Dense(units=4, activation='sigmoid', input_shape=(2,)),
    Dense(units=4, activation='sigmoid'),
    Dense(units=1, activation='sigmoid')
])

model.compile(
   optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X,
    y,
    epochs=5,
    verbose=1
)

print("\nModel Summary:\n")
model.summary()


print("\nMODEL WEIGHTS:\n")

for layer in model.layers:
    weights = layer.get_weights()

    print("Layer:", layer.name)
    print("Weights:\n", weights[0])
    print("Bias:\n", weights[1])
    print("-" * 40)
````
